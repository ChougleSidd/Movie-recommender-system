```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head(5)
```

```python
## to remove space between names to avoid confusion 

movies['genres'] = movies['genres'].apply (lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply (lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply (lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply (lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: x.lower())
```

```python
new.head()
```

```python
## we are using stemming that helps to make root words similar words

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
     
    return " ".join(y)
                       
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
## convert text to vector to find similar word 

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
vector.shape
```

```python
## cosine sim is use as it give the similar recomm movies 

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity =  cosine_similarity(vector)
```

```python
def recommend(movies):
    movies_index = new[new['title'] == movies].index[0]
    distance = similarity[movies_index]
    movies_list  = sorted(list(enumerate(distance)), reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
